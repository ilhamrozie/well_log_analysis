### by7 data ###

```python
well_data = np.genfromtxt("by7.txt").T
depth = well_data[0] # Depth of the measurement in feet
DIR = well_data[1] # Deep induction resistivity measurements in ohm meters
SNR = well_data[2] # Short normal resistivity measurements in ohm meters
SP  = well_data[3] # Spontaneous Potential measurements in millivolts
```

```python
## Make sure the figure is large enough to neatly fit 3 subplots
plt.figure(figsize=(12,8))
plt.suptitle('Well Billie Yates 7', fontsize=16)


## First subplot for the deep induction resistivity data
plt.subplot(121) # 1 row, 3 columns, 1st position
plt.plot(DIR,depth)
plt.plot(SNR,depth)
plt.axvspan(0,497.71,ymin=0.78,ymax=0.8, alpha=0.4, color='green')
plt.axvspan(0,497.71,ymin=0.72,ymax=0.76, alpha=0.4, color='green')
plt.axvspan(0,497.71,ymin=0.49,ymax=0.51, alpha=0.4, color='green')
plt.axvspan(0,497.71,ymin=0.45,ymax=0.48, alpha=0.4, color='green')
plt.axvspan(0,497.71,ymin=0.23,ymax=0.25, alpha=0.4, color='green')
plt.axvspan(0,497.71,ymin=0.2,ymax=0.22, alpha=0.4, color='green')
plt.axvspan(0,497.71,ymin=0.05,ymax=0.15, alpha=0.4, color='green')
## Never forget the title or axes labels!
plt.xlabel(r"Deep Induction Resistivity ($\Omega m$)",fontsize=12)
plt.ylabel(r"Depth (ft)",fontsize=12)
plt.title("Deep Induction resistivity vs. Depth",fontsize=14)
## Invert the y-axis
plt.gca().invert_yaxis()

## Third subplot for the spontaneous potential data
plt.subplot(122) # 1 row, 3 columns, 3rd position
plt.plot(SP,depth)
plt.axvspan(-100,-10,ymin=0.78,ymax=0.8, alpha=0.4, color='yellow')
plt.axvspan(-100,-10,ymin=0.81,ymax=0.83, alpha=0.4, color='yellow')
plt.axvspan(-100,-10,ymin=0.72,ymax=0.76, alpha=0.4, color='yellow')
plt.axvspan(-100,-10,ymin=0.49,ymax=0.51, alpha=0.4, color='yellow')
plt.axvspan(-100,-10,ymin=0.45,ymax=0.48, alpha=0.4, color='yellow')
plt.axvspan(-100,-10,ymin=0.23,ymax=0.25, alpha=0.4, color='yellow')
plt.axvspan(-100,-10,ymin=0.2,ymax=0.22, alpha=0.4, color='yellow')
plt.axvspan(-100,-10,ymin=0.19,ymax=0.22, alpha=0.4, color='yellow')
plt.axvspan(-100,-10,ymin=0.05,ymax=0.15, alpha=0.4, color='yellow')
## Never forget the title or axes labels!
plt.xlabel(r"Spontaneous Potential ($mV$)",fontsize=12)
plt.ylabel(r"Depth (ft)",fontsize=12)
plt.title("Spontaneous Potential vs. Depth",fontsize=14)
## Invert the y-axis
plt.gca().invert_yaxis()

## Make it look nice
plt.tight_layout() 
plt.subplots_adjust(top=0.91)
plt.show()
```

```python
plt.figure(figsize=(5,10))

# Plot each of the three data arrays
plt.plot(DIR,depth,label=r"D.I. Resistivity ($\Omega m$)") # Deep Induction resistivity data
plt.plot(SNR,depth,label=r"S.N. Resistivity ($\Omega m$)") # Short normal resistivity data
plt.plot(SP,depth,label="Spontaneous Potential (mV)") # Spontaneous potential data

## Never forget the title or axes labels!
plt.xlabel(r"Data measurements",fontsize=12)
plt.ylabel(r"Depth (ft)",fontsize=12)
plt.title("Well log data vs. Depth",fontsize=14)
## Invert the y-axis
plt.gca().invert_yaxis()

## We labeled our lines, so we must use a legend
plt.legend()
plt.tight_layout()
plt.show()
```
